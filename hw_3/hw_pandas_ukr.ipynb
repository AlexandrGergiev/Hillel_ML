**4-5. Які средні значення та средньоквадратичні відхилення віку тих, хто отримує більше 50K на рік (ознака *salary*) і тих, хто отримує менше 50K на рік?**

```python
print(f'More than 50K - Mean: {data[data['salary'] == '>50K']['age'].mean()}')
print(f'Standard Deviation: {data[data['salary'] == '>50K']['age'].std()}')
print(f'Less than 50K - Mean: {data[data['salary'] == '<=50K']['age'].mean()}')
print(f'Standard Deviation: {data[data['salary'] == '<=50K']['age'].std()}')
```

**6. Чи правда, що люди, які отримують більше 50k, мають як мінімум вищу освіту? (ознака *education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* или *Doctorate*)**

```python
degree = ['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate', 'Some-college']
rich = data[data['salary'] == '>50K']

rich['education'] = rich['education'].apply(lambda education: education in degree)
have_ed = rich['education'].value_counts(normalize=True)

if have_ed.iloc[0] > have_ed.iloc[1]:
    print(f"Так, більшість багатих людей має вищу освіту: {have_ed.iloc[0]:.2%}")
else:
    print(f'Ні, більшість багатих людей не має вищу освіту: {have_ed.iloc[1]:.2%}')


```

**8. Серед кого більша доля багатіших (>50K): серед одружених чи не одружених чоловіків (ознака *marital-status*)? За одружених вважаємо тих, у кого *marital-status* починається із *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), всі інші - не одружені.**

```python
men = data[data['sex'] == 'Male']

men['marital_status'] = men['marital-status'].apply(lambda status: 'Married' if status.startswith('Married') else 'Not Married')

married_rich = men[men['marital_status'] == 'Married']['salary'].value_counts(normalize=True).get('>50K', 0)

not_married_rich = men[men['marital_status'] == 'Not Married']['salary'].value_counts(normalize=True).get('>50K', 0)

print(f"Багаті серед одружених чоловіків: {married_rich:.2%}")
print(f"Багаті серед не одружених чоловіків: {not_married_rich:.2%}")
if married_rich > not_married_rich:
    print("Серед одружених чоловіків більша доля багатих.")
else:
    print("Серед не одружених чоловіків більша доля багатих.")
```

**9. Яку максимальну кількість годин в тиждень працює людина (ознака *hours-per-week*)? Скільки людей працюють таку кількість годин і який відсоток серед них багатих?**

```python
print(f'Максимальна кількість годин: {data['hours-per-week'].max()}')
print(f'Кількість людей: {data[data['hours-per-week'] == data['hours-per-week'].max()]['hours-per-week'].count()}')

print(f'Відсоток багатих: {data[data['hours-per-week'] == data['hours-per-week'].max()]['salary'].value_counts(normalize=True).get('>50K', 0) * 100}%')
```
