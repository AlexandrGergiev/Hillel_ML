```python
import pandas as pd
import numpy as np
from category_encoders.ordinal import OrdinalEncoder
from category_encoders.sum_coding import SumEncoder
from category_encoders.leave_one_out import LeaveOneOutEncoder
from category_encoders.helmert import HelmertEncoder
from sklearn.preprocessing import OneHotEncoder
from category_encoders.one_hot import OneHotEncoder as OneHotEncoder_ce
from category_encoders.polynomial import PolynomialEncoder
from category_encoders.backward_difference import BackwardDifferenceEncoder
```

```python
data = pd.read_csv('abalone.data', names=['Sex', 'Length', 'Diameter', 'Height', 'Whole weight', 'Shucked weight', 'Viscera weight', 'Shell weight', 'Rings'])
data.head()
```

```python
# onehotenc = OneHotEncoder(handle_unknown='ignore')
# data_ohe = data.join(pd.DataFrame(onehotenc.fit_transform(data[['Sex']]).toarray())).drop(columns=['Sex'])
# # Using sklearn

onehotenc_ce = OneHotEncoder_ce()
data_ohe = data.join(onehotenc_ce.fit_transform(data['Sex']))
#Using category_encoders

data_ohe.head()
```

```python
ord_enc = OrdinalEncoder(cols = ['Sex'])
ord_enc.fit_transform(data)
```

**Признчення натуральних чисел в порядку зустрічі нової змінної. Мінус - відсутність зв'язку з ранжуванням змінної, кожен її вид рівний за сенсом, може вплинути на вагу статі**

```python
LOOE_encoder = LeaveOneOutEncoder()
looe_enc = LOOE_encoder.fit_transform(data[['Sex']], data['Rings'])
looe_enc.head()
```

**Схоже на таргет енкодер, але виключає таргет поточного рядка під час обчислення, щоб зменшити вплив викидів.**

```python
HE_encoder = HelmertEncoder()
helmert_enc = HE_encoder.fit_transform(data[['Sex']], data['Rings']).join(data['Sex'])
helmert_enc.head()
```

**This method creates a new variable which is the difference between the mean of the target for each level and the mean of the target for the previous level.**

```python
poly_enc = PolynomialEncoder(cols = ['Sex'])
poly_enc.fit_transform(data)
```

**Polynomial coding is a form of trend analysis in that it is looking for the linear, quadratic and cubic trends in the categorical variable**

```python
backward_enc = BackwardDifferenceEncoder(cols = ['Sex'])
backward_enc.fit_transform(data)
```

**Cереднє значення фічі для рівня порівнюється із середнім значенням фічі для попереднього рівня**

```python
sum_enc = SumEncoder(['Sex'])
sum_enc.fit_transform(data[['Sex']], data['Rings'])
```

**Подібне до One Hot Encoding, але різниця полягає в тому, що в сумовому кодуванні ми приймаємо одне значення як «-1», і воно не порівнюється з іншим значенням. У той час як у One Hot ми створюємо один стовпець для кожного значення для порівняння з усіма іншими значеннями.**
